## groupby()


#### groupby()와 함께 사용하는 집계 메소드

- count : 누락값을 제외한 데이타 수를 반환
- size : 누락값을 포함한 데이타 수를 반환
- sum
- mean
- std
- var
- min
- max
- quantile(q=0.25)
- quantile(q=0.50)
- quantile(q=0.75)
- describe
- first
- last
- nth   : n번째 행 반환
- sem   : 평균의 표준편차 반환

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
df
```

```python
# 그룹화 - '선별' 단위로 각각의 노선에 대한 총 갯수
df.groupby('선별').count()
    # 각 컬럼으로 동일한 수가 여러번 출력된다.
df_group = df.groupby('선별')    # 우선 특정 항목의 그룹을 다른 변수에 저장
# print(df_group)                # df_group 출력안됨 - DataFrameGroupBy 이라고만 출력
# print(df_group['차종'])        # SeriesGroupBy 이라고 출력

df_group.count()       # 그룹핑된 객체에서 집계함수이용
```

```python
# [주의]
# df_group['차종'].count().sort_values(ascending=0)  # 에러발생!

# 그룹핑하고 한 컬럼에 대한 값만 추출하였다는 것은 Series 데이타 타입이기에
# DataFrame 정렬할 때의 (ascending=0) 방식은 에러 발생. (by=컬럼명)도 사용못함
# df_group['선별'].count().sort_values(ascending=0) # 에러발생

df_group['선별'].count().sort_values(ascending=False)

dd = df_group['선별'].count()
print(type(dd))
```

```python
# 차종별, 노선별별 그룹화 작업
df_group2 = df.groupby(['차종','선별'])
df_group2
type(df_group2)
df_group2.mean()
```

```python
# 교차테이블(cross tab) - 엑셀의 Pivot 유사
result = pd.crosstab(df['차종'], df['선별'])   # 첫번째인자: 인덱스, 두번째인자:컬럼

result2 = pd.crosstab(df['차종'], df['선별'], margins=True)  #  margins=True : All 나오도록
```
