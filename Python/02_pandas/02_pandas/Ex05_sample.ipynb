### 데이타 필터링

# 고속버스 데이타셋에서 인사이트를 도출

[예] 

1. 우승고속에 대한 데이타 필터링(추출)

2. 경부선에서 총이용인원이 1000명 이상인 데이타 필터링(추출)

3. 그 외 데이타를 찾아내시오

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
       # sheetname='highway'  encoding='cp949' 추가 안해도 실행  
df.head()  # 상위 5개만 보임
```

[연습]

```python
# 필터링 - 우등고속에 대한 데이터대상
df_sample = df[df['차종']=='우등']
df_sample
```

```python
# 필터링 - 경부선에서 총이용인원이 1000명 이상인 데이타대상
# 각 조건별로 반드시 ( )로 묶어야 한다
df_sample = df[ (df.선별=='경부선') & (df.총이용인원>=1000)] 
df_sample.sort_values(by='총이용인원', ascending=0).head(10)
```
