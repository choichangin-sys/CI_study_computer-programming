### * 인사이트를 도출하여 정보를 추출하고 시각화

```python
# 대통령 리스트의 정당(Political Party)를 분석
df['Political Party']
```

```python
# 대통령 리스트의 정당(Political Party)를 분석
df['Political Party'].value_counts()
```
