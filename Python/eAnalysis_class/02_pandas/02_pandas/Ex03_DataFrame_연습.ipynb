```python
import pandas as pd

# pd.read_excel('data/2016-01.xls')
df = pd.read_excel('data/2016-01.xls', index_col='상품명')
df.head(10)
```

#### 판다스 데이타프레임의 기본적인 정보 얻어오기

- 앞에 10개 행 보기
- 뒤에 10개 행 보기

- 판매건수에서 큰 값 순으로 10개 (ngargest()이용)
- 판매건수에서 작은 값 순으로 10개 (nsmallest()이용)


```python

df.head(10)
df.tail()

df.nlargest(10, '판매건수') # 큰 값으로 정렬되어 10개 ( 실제 정렬되진 않음 )
df.nsmallest(10,'판매건수')

df.nsmallest(10, ['판매건수','가격'])
```

#### 연속적(chain)으로 데이타 추출

가격순으로 정렬하되 판매건수와 가격만 추출

```python
df.sort_values('가격')
df.sort_values('가격')[['판매건수','가격']]
df.sort_values('가격')[['판매건수','가격']].head()
df.sort_values('가격')[['판매건수','가격']][11:20]
```
